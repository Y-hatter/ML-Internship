## 2.9 

**Plot the 21 day and 34 day Moving average with the average price and decide a Call !
Call should be buy whenever the smaller moving average (21) crosses over longer moving average (34) AND the call should be sell whenever smaller moving average crosses under longer moving average.
One of the most widely used technical indicators.**

```python
# GAIL stocks used for decision
plt.figure(figsize=(20,10))
gail_Date = pd.to_datetime(filter_gail['Date'])
gail_close_price = filter_gail['Close Price']
gailLis=gail_Date.tolist()
gail_change = gail_close_price.pct_change().fillna(0).rolling(7).mean().fillna(0)
plt.plot(gailLis,gail_change,label = 'GAIL')
plt.legend(loc='upper left')
plt.show()
```

```python
import numpy as np
#making small and long signals
smaller_window = 21
longer_window = 34

signals = pd.DataFrame(index=filter_gail.index)
signals['signal'] = 0.0

#SMA of Smaller Window
signals['small_mavg'] = filter_gail['Close Price'].rolling(window=smaller_window, min_periods=1,center=False).mean()

#SMA of Longer Window
signals['long_mavg'] = filter_gail['Close Price'].rolling(window=longer_window,min_periods=1, center=False).mean()

#Create signals
signals['signal'][smaller_window:] = np.where(signals['small_mavg'][smaller_window:] > signals['long_mavg'][smaller_window:], 1.0,0.0)

#Generate trading orders
signals['positions'] = signals['signal'].diff()
print(signals)

```

```python
fig = plt.figure(figsize=(24,16))
ax1 = fig.add_subplot(111, ylabel='Price')
#Plot the closing price
filter_gail['Close Price'].plot(ax=ax1, color='black', lw=2.)

#Plotting the short and long moving averages
signals[['small_mavg', 'long_mavg']].plot(ax=ax1, lw=2.)

#Plotting the buy signals
ax1.plot(signals.loc[signals.positions == 1.0].index, signals.small_mavg[signals.positions == 1.0], '^' , markersize=20,color='g')

#Plotting the sell signals
ax1.plot(signals.loc[signals.positions == -1.0].index, signals.small_mavg[signals.positions == -1.0], 'v' , markersize=20,color='r')

plt.show()
```

## 2.10 
**Trade Calls - Using Bollinger Bands
Plot the bollinger bands for this stock - the duration of 14 days and 2 standard deviations away from the average
The bollinger bands comprise the following data points-
The 14 day rolling mean of the closing price (we call it the average) 
Upper band which is the rolling mean + 2 standard deviations away from the average. 
Lower band which is the rolling mean - 2 standard deviations away from the average. 
Average Daily stock price.
Bollinger bands are extremely reliable , with a 95% accuracy at 2 standard deviations , and especially useful in sideways moving market.
Observe the bands yourself , and analyse the accuracy of all the trade signals provided by the bollinger bands.
Save to a new csv file.**

```python
import pandas as pd
import matplotlib.pyplot as plt

symbol = 'GAIL'



df = pd.read_csv('GAIL.csv'.format(symbol), index_col='Date',
                 parse_dates=True, usecols=['Date', 'Close Price'],
                 na_values='nan')

df = df.rename(columns={'Close Price': symbol})
df.dropna(inplace=True)

# Calculating Simple Moving Average with 14 days window
sma = df.rolling(window=14).mean()

# Calculating the standard deviation
rstd = df.rolling(window=14).std()

upper_band = sma + 2 * rstd
upper_band = upper_band.rename(columns={symbol: 'upper'})
lower_band = sma - 2 * rstd
lower_band = lower_band.rename(columns={symbol: 'lower'})
df = df.join(upper_band).join(lower_band)
ax = df.plot(title='{} Price and BB'.format(symbol))
ax.fill_between(df.index, lower_band['lower'], upper_band['upper'], color='#ADCCFF', alpha='0.3')
ax.set_xlabel('Date')
ax.set_ylabel('SMA and BB')
ax.grid()
plt.show()
```
